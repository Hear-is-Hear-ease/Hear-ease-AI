## 원본 데이터로부터 모델 학습에 사용할 수 있는 데이터를 추출해보자.

프로세싱 순서는 아래와 같다.

1. 사용할 수 없는 데이터는 제외한 다음 상위 7개의 state를 가지는 음성만을 가져온다. 

2. sample rate를 16000으로 통일한다.

3. 로그 멜 스펙트럼을 통해 음성의 파워를 측정한다.

4. 파워의 지역 최솟값을 기준으로 음성을 분할한다.

5. 분할된 음성의 앞뒤 화이트 노이즈(특정 임계값 이하의 에너지를 가지는 시점)을 제거한다.

6. 분할된 음성이 노이즈를 제거하여 패턴을 보다 잘 드러나게 한다.

7. 음성 정규화를 수행한다.

8. Padding과 Trimming을 통해 평균 세그멘테이션 길이를 가지도록 음성의 길이를 통일한다.

9. 음성의 유사도를 측정하여 하위 10%의 데이터는 제외한다.

10. 유사도 상위 10% 데이터에 대하여 SpecAugment를 적용하여 Training 데이터에 대해 마스킹 작업을 수행한다.

Dilated Convolution: 이 방식은 컨볼루션 연산 시 간격을 두어 더 넓은 범위의 정보를 포착할 수 있게 합니다. 이를 통해 깊은 네트워크에서도 좀 더 넓은 시간 범위의 정보를 유지할 수 있습니다.

Residual Connections (Skip Connections): ResNet 아키텍처에서 사용되는 방식으로, 이전 레이어의 출력을 뒷단의 레이어 입력에 더해줌으로써 정보의 손실을 최소화합니다.

Attention Mechanisms: 특히 음성 인식과 같은 시퀀스 데이터에서, 어텐션 메커니즘을 통해 중요한 시간적 특징에 집중할 수 있게 합니다.
